# PASO 1.1 — Cargar y unir los datasets


```python
# Cargar archivos
# ya que los datos ya estan cargados anteriormente solo corremos los datos

# Unificación por customerID
df = (
    contract
    .merge(personal, on="customerID", how="left")
    .merge(internet, on="customerID", how="left")
    .merge(phone, on="customerID", how="left")
)

# Vista general
df.head()
```

```python
# PASO 1.2.A — Corrección de tipos 

# Convertir fechas
if "BeginDate" in df.columns:
    df["BeginDate"] = pd.to_datetime(df["BeginDate"], format='%Y-%m-%d', errors="coerce")

if "EndDate" in df.columns:
    df["EndDate"] = pd.to_datetime(df["EndDate"], format='%Y-%m-%d', errors="coerce")

# Convertir columnas numéricas
numeric_cols = ["MonthlyCharges", "TotalCharges"]

for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
# PASO 1.2.B — Rellenar valores faltantes en servicios

internet_cols = [
    "InternetService",
    "OnlineSecurity",
    "OnlineBackup",
    "DeviceProtection",
    "TechSupport",
    "StreamingTV",
    "StreamingMovies"
]

for col in internet_cols:
    if col in df.columns:
        df[col] = df[col].fillna("No")


# Para phone.csv
if "MultipleLines" in df.columns:
    df["MultipleLines"] = df["MultipleLines"].fillna("No")
```

```python
# PASO 1.2.C — Limpieza de TotalCharges

if "TotalCharges" in df.columns:
    df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
    df["TotalCharges"] = df["TotalCharges"].fillna(0)
```

```python
# PASO 3.2 — Contar servicios contratados

service_cols = [
    "OnlineSecurity", "OnlineBackup", "DeviceProtection",
    "TechSupport", "StreamingTV", "StreamingMovies", "MultipleLines"
]

df["NumServices"] = df[service_cols].apply(lambda row: sum(row == "Yes"), axis=1)
```

```python
# PASO 3.3 — Indicadores binarios

df["HasInternet"] = (df["InternetService"] != "No").astype(int)
df["PaperlessBillingFlag"] = (df["PaperlessBilling"] == "Yes").astype(int)
```

## Resumen

### Contexto y objetivo estratégico
Interconnect, operador de telecomunicaciones con servicios de telefonía fija e Internet, enfrenta una tasa significativa de cancelación de clientes (churn). La pérdida de usuarios impacta directamente los ingresos recurrentes y aumenta los costos de adquisición de nuevos clientes. El objetivo del proyecto fue desarrollar un modelo predictivo capaz de identificar clientes con alta probabilidad de cancelar su servicio, permitiendo al equipo de marketing implementar acciones preventivas como ofertas personalizadas, descuentos o migración a planes más estables.


### Metodología aplicada
El proyecto se desarrolló siguiendo un flujo completo de ciencia de datos:
- Integración de datos provenientes de cuatro fuentes: contrato, datos personales, servicios de Internet y servicios telefónicos.
- Limpieza y estandarización, asegurando consistencia en tipos de datos, manejo de valores faltantes y coherencia entre tablas.
- Análisis exploratorio (EDA) para identificar patrones de comportamiento asociados al churn.
- Ingeniería de características, incluyendo variables como antigüedad del cliente (tenure), número de servicios contratados y banderas de comportamiento.
- Modelado supervisado, evaluando múltiples algoritmos: Regresión Logística, Árboles de Decisión, Random Forest, SVM, LightGBM y XGBoost.
- Optimización y evaluación, priorizando métricas críticas para el negocio como Recall, F1-score y AUC‑ROC.
- Construcción de un dashboard interactivo para visualizar métricas clave del modelo.


### Resultados principales
El modelo con mejor desempeño fue XGBoost, logrando:
- Recall: 0.89
Capacidad sobresaliente para identificar clientes que efectivamente cancelan.
Esta métrica es crítica, ya que cada cliente no detectado implica pérdida directa de ingresos.
- AUC‑ROC: 0.83
Indica una buena capacidad de discriminación entre clientes que cancelan y los que permanecen.
- F1-score equilibrado, adecuado para un dataset desbalanceado.
El modelo final fue interpretado mediante técnicas de importancia de variables y SHAP, garantizando transparencia y explicabilidad para toma de decisiones.


### Principales factores que impulsan el churn
El análisis reveló patrones consistentes con la industria de telecomunicaciones:
- Tipo de contrato: los contratos mensuales presentan la mayor tasa de churn; los contratos de 1 y 2 años son significativamente más estables.
- Antigüedad del cliente (Tenure): los clientes nuevos son más propensos a cancelar.
- Cargos mensuales elevados: clientes con facturas altas muestran mayor sensibilidad al precio.
- Método de pago: Electronic Check se asocia a mayor churn; pagos automáticos reducen la probabilidad de cancelación.
- Servicios de soporte y seguridad: la ausencia de OnlineSecurity o TechSupport incrementa el riesgo de churn.
- Tipo de Internet: los usuarios de fibra óptica presentan mayor churn, posiblemente por expectativas más altas o costos superiores.


### Recomendaciones de negocio
A partir de los hallazgos, se proponen acciones concretas:
- Campañas de retención dirigidas a clientes con contrato mensual, especialmente aquellos con cargos altos y baja antigüedad.
- Incentivos para migrar a contratos anuales, como descuentos por permanencia o beneficios exclusivos.
- Promoción de servicios de soporte y seguridad, que demostraron reducir churn.
- Revisión de precios y beneficios para clientes de fibra óptica, segmento con mayor riesgo.
- Fomento de métodos de pago automáticos, mediante descuentos o puntos de fidelidad.
- Implementación de un sistema de alertas tempranas basado en el modelo, para priorizar clientes con mayor probabilidad de cancelar.


### Impacto esperado
La adopción del modelo permitirá:
- Reducir la tasa de churn mediante intervenciones proactivas.
- Optimizar el presupuesto de marketing enfocando esfuerzos en clientes con mayor riesgo.
- Aumentar la retención y el valor de vida del cliente (Customer Lifetime Value).
- Mejorar la experiencia del cliente mediante ofertas personalizadas y oportunas.
