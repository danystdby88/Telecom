#### A continuación, se detalla la hoja de ruta estratégica para el desarrollo del proyecto:

1. Integración y Limpieza de Datos

1.1 Unificación de fuentes: Creación de un dataset maestro a partir de las tablas disponibles.

1.2 Tratamiento de datos: Gestión de valores faltantes y corrección de tipos de variables.

2. Análisis Exploratorio de Datos (EDA)

Identificación de patrones de comportamiento, correlaciones y detección de valores atípicos (outliers).

3. Preparación de Datos para Modelado

Ingeniería de variables (Feature Engineering), codificación de variables categóricas y normalización.

4. Selección y Entrenamiento de Modelos

Se evaluarán distintos algoritmos para comparar su desempeño:

Regresión Logística (como baseline interpretable).

Modelos de ensamble: Random Forest y Gradient Boosting (XGBoost, LightGBM o CatBoost).

Árboles de Decisión y SVM (según el volumen de datos).

Redes Neuronales simples (opcional).

5. Ajuste de Hiperparámetros y Evaluación

Optimización de modelos para maximizar métricas clave:

Recall: Crucial para identificar a la mayor cantidad de clientes en riesgo.

F1-Score: Balance entre precisión y sensibilidad.

AUC-ROC y Matriz de Confusión.

6. Implementación y Entregables

Dataset: Datos limpios y debidamente documentados.
